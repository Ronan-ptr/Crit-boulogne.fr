```python
import pandas as pd
import os
WRK_DIR = "/home/prime/Projets/2026_PROJECT-DA/DATA/EVENTS" #Directoire des fichier evenemenets
```

```python
files_names = [] # List of all files names in the directory
files_shape = [] # List of all shapes for each file
files_col = [] # List of all colums name for one files
files_col_nb = [] # List of all numbers of colums for each file

# LOOP FOR EACH FILES
for file in os.listdir(WRK_DIR) : # for each files in the path
  df = pd.read_excel(WRK_DIR + "/" + file) # read the file
  print(f"File {file} is reading") # Debug
  
  files_names.append(file) # Add file name
  files_shape.append(df.shape) # Add file shape
  files_col_nb.append(len(df.columns)) # add number of col

  col = list(df.columns)
  files_col.append(col) # Add list of columns names

# DataFrame creation form a dictionnary
files_infos = {
    "names" : files_names,
    "shape" : files_shape,
    "col_nb" : files_col_nb,
    "col_names" : files_col
}

infos = pd.DataFrame(files_infos)
infos.head() # See df
```

```python
events = pd.DataFrame() # Crée un df vide

for file in os.listdir(WRK_DIR) : # for each files in the path
  df = pd.read_excel(WRK_DIR + "/" + file) # read the file
  print(f"File {file} is reading") # Debug

  df["Event"] = str(file) # Nouvelle colonne nomme par le nom du fichier dans chaque celule

  events = pd.concat([events, df]) # Concatene les df

#display(events.shape)
events.head()
```

```python
events.to_csv("/home/prime/Projets/2026_PROJECT-DA/DATA/events.csv")
```

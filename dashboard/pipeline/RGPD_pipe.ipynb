```python
import pandas as pd 

events = pd.read_csv("/home/prime/Projets/2026_PROJECT-DA/DATA/events.csv")
pax = pd.read_csv("/home/prime/Projets/2026_PROJECT-DA/DATA/pax.csv")
```

#### INFORMATIONS

#### FUSION
Fusionner les DataFrame dans un nouveaux
On vas certainemetn garder l'ensemble des evenement et fuisionner les pax dessus

```python
dataset = events.merge(pax, on="Matricule", how="left")
dataset = events.merge(pax, on="Matricule", how="inner")
dataset.head()
```
